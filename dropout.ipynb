```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv("C:\\Users\\User\\Downloads\\dataset.csv")
df.head()
```

```python

print(df.shape)
print(df.columns)
# print(df.describe())

# print(df.isnull().sum())
```

```python
lb=LabelEncoder()
df["Target"]=lb.fit_transform(df["Target"])
```

```python
y=df["Target"]
x = df[[ 'Application mode', 'Application order', 'Course', 
        'Daytime/evening attendance', 'Displaced', 'Educational special needs', 
        'Debtor', 'Tuition fees up to date', 'Scholarship holder']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
model = LogisticRegression(max_iter=1000)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
acc = accuracy_score(y_test,y_pred)
print("Accuracy:",round(acc*100,2),"%")
```
